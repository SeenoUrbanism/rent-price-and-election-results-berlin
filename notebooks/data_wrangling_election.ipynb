```python
def aggregate_bezirk(df):
    # Define which columns are numeric vote counts
    parties = ['spd', 'cdu', 'gruene', 'fdp', 'afd', 'die_linke']
    
    # Compute totals by Bezirk
    grouped = df.groupby('bezirksname', as_index=False).agg({
        'wahlberechtigte_insgesamt': 'sum',
        'waehler': 'sum',
        'gueltige_stimmen': 'sum',
        **{p: 'sum' for p in parties}
    })
    
    # Recalculate percentages from these totals
    for p in parties:
        grouped[p + '_%'] = (grouped[p] / grouped['gueltige_stimmen']) * 100
    
    grouped['wahlbeteiligung_%'] = (grouped['waehler'] / grouped['wahlberechtigte_insgesamt']) * 100
    
    return grouped
```

```python
df_2017_agg = aggregate_bezirk(df_2017)
combined_2021_agg = aggregate_bezirk(combined_2021)
df_2025_agg = aggregate_bezirk(df_2025)
```

```python
def aggregate_by_bezirk(df):
    # Define relevant vote count columns
    parties = ['spd', 'cdu', 'gruene', 'fdp', 'afd', 'die_linke']
    
    # Group by district and sum all numeric columns
    agg_df = df.groupby('bezirksname', as_index=False).agg({
        'wahlberechtigte_insgesamt': 'sum',
        'waehler': 'sum',
        'gueltige_stimmen': 'sum',
        **{p: 'sum' for p in parties}
    })
    
    # Calculate turnout (%)
    agg_df['wahlbeteiligung_%'] = (agg_df['waehler'] / agg_df['wahlberechtigte_insgesamt']) * 100

    # Calculate party percentages (share of valid votes)
    for p in parties:
        agg_df[p + '_%'] = (agg_df[p] / agg_df['gueltige_stimmen']) * 100
    # Calculate "others" percentage
    major_parties_pct = [p + '_%' for p in parties]
    agg_df['others_%'] = 100 - agg_df[major_parties_pct].sum(axis=1)
    
    # Round all numeric columns to two decimals
    numeric_cols = agg_df.select_dtypes(include=['float64', 'int64']).columns
    agg_df[numeric_cols] = agg_df[numeric_cols].round(2)
    
    return agg_df
```

```python
df_2017_agg = aggregate_by_bezirk(df_2017)
combined_2021_agg = aggregate_by_bezirk(combined_2021)
df_2025_agg = aggregate_by_bezirk(df_2025)
```
